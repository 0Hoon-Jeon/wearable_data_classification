```python
import os
import glob
import shutil
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
```

```python
dataset_path = '/home/ubuntu/Dataset/200630/wearable_dataset/'
prep_path = '/home/ubuntu/Dataset/prep_path_pulse1/'
```

```python
location_format =  ['ToeL_DCD4_Calibrated_PC','ToeR_DCAF_Calibrated_PC',
                    'LegL_DDCC_Calibrated_PC', 'LegR_DC90_Calibrated_PC',
                    'ThighL_DC74_Calibrated_PC', 'ThighR_DC77_Calibrated_PC', 
                    'ThighR2_A640_Calibrated_PC','Waist_E0E6_Calibrated_PC']
```

```python
def overview_dataset(root_path):
    s_list = os.listdir(dataset_path)
    s_list.sort()
    print('subject_num = {}\n'.format(len(s_list)))
    print('s_list : ',s_list)
    
    cnt = np.zeros(shape=(7))    
    
    tmp_dict = dict()
    
    format_list = []
    
    
    for s in s_list:
        tmp_dict[s]=dict()
        for i in range(1,8):
            tmp_dict[s]['E{}'.format(i)]=[]
        s_num = int(s[1:])
        subroot_path = root_path+s+'/'
        
        for (path, dir, files) in os.walk(subroot_path):
            for filename in files:
                if 'Un' not in filename and '.csv' in filename:
                    tmp = filename.split('_')
                    e = tmp[0][-2:]
                    loc = filename[filename.index('_',10)+1:filename.index('.')]
                    if loc not in format_list:
                        format_list.append(loc)
                    tmp_dict[s][e].append(filename)
        
        
        print('\ns_num : ',s_num,'subroot_path : ',subroot_path)
        for e_num in range(1,8):
            cnt[e_num-1] += len(tmp_dict[s]['E{}'.format(e_num)])/7
            print('E{} :'.format(e_num),len(tmp_dict[s]['E{}'.format(e_num)]),end=' | ')
        print()
    print('\nlocation_format : ',format_list)
    print('cnt : ', cnt)
    print('\nExample')
    for filename in tmp_dict['S180']['E2']:
        print(filename)
                        
            
```

```python
def get_list(root_path):
    s_list = os.listdir(dataset_path)
    s_list.sort()   
    
    tmp_dict = dict()
    
    for s in s_list:
        tmp_dict[s]=dict()
        for i in range(1,8):
            tmp_dict[s]['E{}'.format(i)]=[]
        s_num = int(s[1:])
        subroot_path = root_path+s+'/'
        
        for (path, dir, files) in os.walk(subroot_path):
            for filename in files:
                if 'Un' not in filename and 'Waist' in filename and '.csv' in filename:
                    tmp = filename.split('_')
                    e = tmp[0][-2:]
                    loc = filename[filename.index('_',10)+1:filename.index('.')]
                    tmp_dict[s][e].append(path+'/'+filename)
    
    return tmp_dict
            
```

```python

def mkdir(prep_path,tmp_dict):
    try:
        os.mkdir(prep_path)
    except:
        print('Already exist, delete and end')
        shutil.rmtree(prep_path)
        return False
    for s_num in tmp_dict:
        os.mkdir(prep_path+s_num)
        for e in range(1,8):
            e_num = 'E{}'.format(e)
            os.mkdir(prep_path+s_num+'/'+e_num)
```

```python
def sel_col(df, col_name):
    res = ''
    for column in df.columns:
        if col_name in column:
            res = column
            break
    if res == '':
        
        print(df.columns,col_name,'not existed')
        return
    else:
        return res 
def load_data(filename):
    df = pd.read_csv(filename , skiprows=[0,2], sep='	')
    col_list = ['Timestamp', 'Accel_WR_X','Accel_WR_Y','Accel_WR_Z' , 'Gyro_X','Gyro_Y','Gyro_Z' , 'Mag_X','Mag_Y','Mag_Z','Marker']
    
    data_dict= dict()
    
    for col in col_list :
        data_dict[col] = list(df[sel_col(df,col)])

    
    
    try:
        T1_stind = data_dict[sel_col(df,'Marker')].index(2)
        T1_endind= T1_stind + data_dict[sel_col(df,'Marker')][T1_stind:].index(0) -1
    except:
        T1_stind,T1_endind = None,None
    
    try:
        T2_stind= T1_endind +1 + data_dict[sel_col(df,'Marker')][T1_endind+1:].index(2) 
        T2_endind= T2_stind + data_dict[sel_col(df,'Marker')][T2_stind:].index(0) -1
    except:
        T2_stind,T2_endind = None,None
    
    data_dict['pulse']=[T1_stind,T1_endind,T2_stind,T2_endind]
    return data_dict
```

```python
tmp_dict = get_list(dataset_path)
tmp = mkdir(prep_path,tmp_dict)
if tmp == False:
    print('hi')
    mkdir(prep_path,tmp_dict)
for s_num in tmp_dict:
    print(s_num)
    e_list=['E{}'.format(x) for x in range(1,5)]
    for e_num in e_list:
#     for e_num in tmp_dict[s_num]:
        for index,waist_filename in enumerate(tmp_dict[s_num][e_num]):
            
            Waist_filename = waist_filename
            ThighR_filename = waist_filename.replace(location_format[-1],location_format[-3])
            ThighL_filename = waist_filename.replace(location_format[-1],location_format[-4])
            LegR_filename = waist_filename.replace(location_format[-1],location_format[-5])
            LegL_filename = waist_filename.replace(location_format[-1],location_format[-6])
            ToeR_filename = waist_filename.replace(location_format[-1],location_format[-7])
            ToeL_filename = waist_filename.replace(location_format[-1],location_format[-8])
            
            try:
                Waist_data = load_data(Waist_filename)
                try:
                    ThighR_data = load_data(ThighR_filename)
                except:
                    ThighR_filename = waist_filename.replace(location_format[-1],location_format[-2])
                    ThighR_data = load_data(ThighR_filename)

                ThighL_data = load_data(ThighL_filename)
                LegR_data = load_data(LegR_filename)
                LegL_data = load_data(LegL_filename)
                ToeR_data = load_data(ToeR_filename)
                ToeL_data = load_data(ToeL_filename)
            except:
                print(Waist_filename)
                continue
            
            data = [ Waist_data,ThighR_data,ThighL_data,LegR_data,LegL_data,ToeR_data,ToeL_data]
            data_len = [len(x['Timestamp']) for x in data]
#             print(data_len,min(data_len))
            
            col_list = [ 'Accel_WR_X','Accel_WR_Y','Accel_WR_Z' , 'Gyro_X','Gyro_Y','Gyro_Z' , 'Mag_X','Mag_Y','Mag_Z']
        
            tmp = np.zeros((min(data_len),7*9))
            
            
#             plt.figure(fig_size=(15,15))
            for j,loc_data in enumerate(data):
                for i,a in enumerate(col_list):
                    t = signal.resample(loc_data[a][loc_data['pulse'][0]:loc_data['pulse'][1]],min(data_len))
                    tmp[:,j*9+i] = t
#             print(tmp.shape)        
            
            save_filename = '{}.npy'.format(index)
            save_path = prep_path+'{}/{}/'.format(s_num,e_num)
#             print(save_path+save_filename)
            np.save(save_path+save_filename,tmp)
```
